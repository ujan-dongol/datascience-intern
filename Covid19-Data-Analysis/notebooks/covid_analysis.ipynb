# COVID-19 Data Analysis

This project analyzes global COVID-19 data to understand trends in confirmed cases, recoveries, and deaths.

The goal is to explore patterns and visualize how the pandemic evolved across different countries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/covid_data.csv")

df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
global_cases = df.groupby("Date")["Confirmed"].sum()

plt.figure(figsize=(10,5))

plt.plot(global_cases)

plt.title("Global COVID-19 Confirmed Cases Over Time")

plt.xlabel("Date")
plt.ylabel("Cases")

plt.savefig("../images/global_cases.png")

plt.show()
```

```python
global_deaths = df.groupby("Date")["Deaths"].sum()

plt.figure(figsize=(10,5))

plt.plot(global_deaths)

plt.title("Global COVID-19 Deaths Over Time")

plt.xlabel("Date")
plt.ylabel("Deaths")

plt.savefig("../images/global_deaths.png")

plt.show()
```

```python
global_recovered = df.groupby("Date")["Recovered"].sum()

plt.figure(figsize=(12,6))
plt.plot(global_recovered)

plt.title("Global Recovery Trend")
plt.xlabel("Date")
plt.ylabel("Recovered")

plt.savefig("../images/global_recovery_trend.png")

plt.show()
```

```python
latest = df[df["Date"] == df["Date"].max()]

top_countries = latest.sort_values("Confirmed", ascending=False).head(10)

plt.figure(figsize=(10,5))

sns.barplot(x="Confirmed", y="Country", data=top_countries)

plt.title("Top 10 Countries by Confirmed Cases")

plt.savefig("../images/top_countries_cases.png")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.scatterplot(x="Confirmed", y="Recovered", data=latest)

plt.title("Confirmed vs Recovered Cases")

plt.savefig("../images/recovery_analysis.png")

plt.show()
```

```python
top_deaths = latest.sort_values("Deaths", ascending=False).head(10)

plt.figure(figsize=(10,6))

sns.barplot(x="Deaths", y="Country", data=top_deaths)

plt.title("Top 10 Countries by Deaths")

plt.savefig("../images/top10_deaths.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(x="Confirmed", y="Deaths", data=latest)

plt.title("Confirmed vs Deaths Relationship")

plt.savefig("../images/confirmed_vs_deaths.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(x="Confirmed", y="Recovered", data=latest)

plt.title("Confirmed vs Recovered")

plt.savefig("../images/confirmed_vs_recovered.png")

plt.show()
```

```python
plt.figure(figsize=(8,6))

corr = latest[["Confirmed","Deaths","Recovered"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Between COVID Metrics")

plt.savefig("../images/correlation_heatmap.png")

plt.show()
```

## Conclusion

The analysis shows how COVID-19 cases and deaths evolved over time across the world.

Visualizations highlight the rapid growth of cases during peak periods and reveal which countries were most affected.
